```python
import torch
```

```python
random_tensor = torch.rand(5)
print(random_tensor)
print(random_tensor.ndim)
```

```python
torch.manual_seed(1) #to get same values
rt = torch.rand(2, 5, 3) #creates a 3-dimensional tensor
'''
2 blocks

Each block has 5 rows

Each row has 3 columns
'''
print(rt)
print(rt.ndim)
print(rt.shape)
```

```python
#torch.randint(low, hight, size)
rt = torch.randint(0, 10, (5, 3))
print(rt)
print(rt.ndim)
print(rt.shape)
```

```python
#generate random number for normal distribution
# torch.normal(mean, std, size)
rt = torch.normal(0.5, 2, (5,3))
print(rt)
print(rt.ndim)
print(rt.shape)

# plt.imshow(rt, cmap='grey')
# plt.title("Uniform Noise Image")
# plt.show()
```

## create noise image

```python
import matplotlib.pyplot as plt
t = torch.rand((256,256))
plt.imshow(t, cmap='grey')
plt.axis('off')
plt.title("Uniform Noise Image")
plt.show()
```

```python
#rgb noise
t_rgb = torch.rand((256,256, 3 ))
plt.imshow(t_rgb.numpy(), cmap='grey')
plt.axis('off')
plt.title("Uniform Noise Image")
plt.show()
```

```python
print(t_rgb.shape)
print(t_rgb.dtype)
print(torch.min(t_rgb))
print(torch.max(t_rgb))
```

## zero and ones

```python
zt = torch.zeros(5)
print(zt)

zt2d = torch.zeros(256, 256)
print(zt2d)
print(zt2d.ndim)
print(zt2d.shape)

plt.imshow(zt2d.numpy(), cmap='grey', vmin = 0 , vmax=1) #vmin = value min
#0 - black 1- white
plt.axis('off')
plt.title("Zero Noise Image")
plt.show()
```

```python
zt2d = torch.ones(256, 256)
print(zt2d)
print(zt2d.ndim)
print(zt2d.shape)

plt.imshow(zt2d.numpy(), cmap='grey', vmin = 0 , vmax=1) #vmin = value min
#0 - black 1- white
#plt.axis('off')
plt.title("white Image")

plt.show()
```

```python
rt = torch.randint(0, 2, (5, 3))

plt.imshow(rt.numpy(), cmap='grey', vmin = 0 , vmax=1) #vmin = value min
#0 - black 1- white
#plt.axis('off')
plt.title("white Image")

plt.show()
```

```python
r = torch.arange(12).unsqueeze(1) #tensor.unsqueeze(dim)
c = torch.arange(12).unsqueeze(0)

b = (r+c) %2

plt.imshow(b.numpy(), cmap='grey', vmin = 0 , vmax=1) #vmin = value min
#0 - black 1- white
#plt.axis('off')
plt.title("zero and one Image")

plt.show()


```
